```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv("data/record-[2018.11.16-10.16.29].csv");
data = raw_data.drop(["Event Id",
                      "Event Date",
                      "Event Duration",
                      "Epoch",
                      "Channel 1",
                      "Channel 4",
                      "Channel 5",
                      "Channel 9",
                      "Channel 12",
                      "Channel 13",
                      "Channel 16",
                      "Time:512Hz"], axis=1)
del raw_data;
fs_data = 512;

blink_data = pd.read_csv('matlab/sub12Trial02.csv', header=None);
fs_blink = 256;
```

```python
[sum(~np.isfinite(data).iloc[:,i]) for i in range(0,16)]
```

```python
data[~data.isin([np.nan, np.inf, -np.inf]).any(1)]
data.head()
```

```python
blink_data.head()
```

```python
from scipy.signal import decimate
r = int(np.ceil(fs_data/fs_blink));
fs = fs_data/r;
ts = 1/fs;

signal_data = decimate(data.T, r, ftype='fir', n=2000, zero_phase=True);
comp_data = blink_data.T.values

if(len(signal_data[1,:]) != len(comp_data[1,:])):
    if(len(signal_data[1,:]) > len(comp_data[1,:])):
        n = len(comp_data[1,:]);
        signal_data = signal_data[:, 0:n]
    else:
        n = len(signal_data[1,:]);
        comp_data = comp_data[:, 0:n]

t = np.arange(0,n)/fs;
```

```python
signal_data
```

```python
from sklearn.decomposition import FastICA

ica = FastICA(fun='exp')
S = ica.fit_transform(list(map(list, zip(*signal_data))))
signal_ica = list(map(list, zip(*S)))
```

```python
plt.plot(signal_ica[2])
```

```python
import pywt

pywt.scale2frequency('morl', [10, 800])*2*fs
```

```python
coef_ica, freqs = pywt.cwt(signal_ica[2], np.arange(5,450), 'morl', sampling_period=ts)
```

```python
coef_sig, freqs = pywt.cwt(signal_data[2], np.arange(5,450), 'morl', sampling_period=ts)
```

```python
coef_blk, freqs = pywt.cwt(comp_data[0,:], np.arange(5,450), 'morl', sampling_period=ts)
```

```python
fwindow = [[0.5, 35], [7,35], [4,7], [0.5, 4]];
twindow = [0, 15];

fmask = [[x >= f[0] and x < f[1] for x in freqs] for f in fwindow]
tmask = [x >= twindow[0] and x < twindow[1] for x in t]
```

```python
for i in range(0, len(fwindow)):
    plt.subplot(len(fwindow), 1, i+1)
    plt.pcolormesh(t[tmask], freqs[fmask[i]], coef_blb[np.ix_(fmask[i],tmask)])
    plt.show();
```

```python
for i in range(0, len(fwindow)):
    plt.subplot(len(fwindow), 1, i+1)
    plt.pcolormesh(t[tmask], freqs[fmask[i]], coef_sig[np.ix_(fmask[i],tmask)])
    plt.show();
```
